```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("diabetes dataset real.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
input_data = dataset.iloc[:,:-1]
output_data = dataset["Outcome"]
```

```python
output_data.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
input_data ,output_data = ro.fit_resample(input_data,output_data)
```

```python
output_data.value_counts()
```

```python
input_data.head()
```

```python
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)
```

```python
x_train , x_test , y_train , y_test = train_test_split(input_data,output_data,test_size=0.2 , random_state=42)
```

```python
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
x_train.shape[1]
```

```python
# Set random seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Model architecture and training (same as previous)
model = Sequential([
    Dense(32, activation='relu', input_dim=8),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dropout(0.3),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
from tensorflow.keras.regularizers import l2

model = Sequential([
    Dense(32, activation='relu', input_dim=8, kernel_regularizer=l2(0.01)),  # L2 Regularization
    Dropout(0.3),
    Dense(16, activation='relu', kernel_regularizer=l2(0.01)),
    Dropout(0.3),
    Dense(8, activation='relu', kernel_regularizer=l2(0.01)),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
# Model Summary to see the architecture
model.summary()
```

```python
# Train the model with early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', patience=5, factor=0.5)
```

```python
history = model.fit(x_train, y_train, epochs=100, batch_size=64, validation_data=(x_test, y_test), callbacks=[early_stopping, lr_scheduler])
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {test_accuracy * 100}%')
```

```python
def predict_diabetes(new_data, scaler, model):
    # Convert new data into DataFrame
    new_data = pd.DataFrame([new_data], columns=input_data.columns)

    # Standardize the input using the same scaler
    new_data_scaled = scaler.transform(new_data)

    # Predict the probability
    prediction = model.predict(new_data_scaled)

    # Convert to 0 or 1 (Threshold = 0.5)
    return int(prediction[0] > 0.5)
```

```python
# Example new data (replace with actual values)
new_patient = [8,	183,	64,	0,	0,	23.3,	0.672,	32]  # 8 features
result = predict_diabetes(new_patient, ss, model)

print(f'Prediction: {result}')  # Output: 0 (No Diabetes) or 1 (Diabetes)
```
